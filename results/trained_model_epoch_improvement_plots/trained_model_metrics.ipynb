```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results.csv')
```

```python
plt.style.use('seaborn-v0_8-deep')

# Function to create and save a plot
def create_plot(x, y, title, xlabel, ylabel, filename):
    plt.figure(figsize=(12, 6))
    for col in y:
        plt.plot(df[x], df[col], label=col)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()
```

```python
create_plot('epoch', 
            ['train/box_loss', 'train/cls_loss', 'train/dfl_loss', 
             'val/box_loss', 'val/cls_loss', 'val/dfl_loss'],
            'Training and Validation Losses', 'Epoch', 'Loss', 'loss_curves.png')
```

```python
create_plot('epoch', 
            ['metrics/precision(B)', 'metrics/recall(B)', 
             'metrics/mAP50(B)', 'metrics/mAP50-95(B)'],
            'Model Performance Metrics', 'Epoch', 'Metric Value', 'metrics_graph.png')
```

```python
create_plot('epoch', ['lr/pg0', 'lr/pg1', 'lr/pg2'],
            'Learning Rate Schedule', 'Epoch', 'Learning Rate', 'lr_schedule.png')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['epoch'], df['time'])
plt.title('Training Time per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Time')
plt.tight_layout()
plt.savefig('training_time.png')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP50')
plt.plot(df['epoch'], df['metrics/mAP50-95(B)'], label='mAP50-95')
plt.title('mAP50 vs mAP50-95')
plt.xlabel('Epoch')
plt.ylabel('mAP')
plt.legend()
plt.tight_layout()
plt.savefig('map_comparison.png')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['train/box_loss'], df['train/cls_loss'])
plt.title('Box Loss vs Classification Loss')
plt.xlabel('Box Loss')
plt.ylabel('Classification Loss')
plt.tight_layout()
plt.savefig('box_vs_cls_loss.png')
plt.show()
```
