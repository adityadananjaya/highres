```python
from transformers import pipeline
from hugging_face_helper_function import helper
```

```python
def add_model(boat_detectors, str, model_url):
    boat_detector = pipeline(str, model=model_url)
    boat_detectors.append(boat_detector)
```

```python
str_detect = "object-detection"

# Load models
boat_detectors = []

add_model(boat_detectors, str_detect, "nielsr/detr-finetuned-boat-detection")
add_model(boat_detectors, str_detect, "zhuchi76/detr-resnet-50-finetuned-boat-dataset")
add_model(boat_detectors, str_detect, "wennnny/detr-resnet-50-finetuned-real-boat-dataset")
add_model(boat_detectors, str_detect, "leowang707/detr-resnet-50-finetuned-real-boat-dataset")
add_model(boat_detectors, str_detect, "facebook/detr-resnet-50")
```

```python
# Run models
images = ["test_resolution_images/boats-16mp.jpg", "test_resolution_images/test8-64mp.jpg"]
image_set = 1

for model in boat_detectors:
    for image in images:
        dest = "model_output_images/" + str(image_set) + "-" + image[image.index("/") + 1:]

        helper(model, image, dest)

    image_set += 1
    
```
